```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

import math
```

```python
dfRes1 = pd.read_csv('../data/interim/Methylation_windows_breast.csv')
dfRes2 = pd.read_csv('../data/interim/Methylation_windows_thyroid.csv')
dfRes3 = pd.read_csv('../data/interim/Methylation_windows_endometrium.csv')
dfRes4 = pd.read_csv('../data/interim/Methylation_windows_kidney.csv')
dfRes5 = pd.read_csv('../data/interim/Methylation_windows_prostate.csv')
```

```python
dfAll = pd.concat((dfRes1, dfRes2, dfRes3, dfRes4, dfRes5))
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.boxplot(x="CHROMOSOME", y="METHYLATION (H)", hue="PRIMARY_SITE", data=dfAll, palette="Set3", ax=ax)
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
plt.savefig('../plots/BoxPlot.png')
```
